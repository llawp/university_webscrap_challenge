```python
# import neighborhood data from source URL
url = "https://www.4icu.org/reviews/index0001.htm"
data = requests.get(url).text 
table = BeautifulSoup(data, 'html5lib')
tables = table.find_all('table')
tables[0].prettify()
```

```python
# make empty table
df = pd.DataFrame(columns=["University Name", "Country code"])

# convert html into table data

for row in tables[0].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        univ = col[0].text.lstrip(" ")
        country = col[1].text.rstrip(" ")
        df = df.append({"University Name":univ, "Country code":country}, ignore_index=True)

df
```

```python
# import neighborhood data from source URL
url = "https://www.iban.com/country-codes"
data = requests.get(url).text 
table = BeautifulSoup(data, 'html5lib')
tables = table.find_all('table')
tables[0].prettify()

# make empty table
df_cc = pd.DataFrame(columns=["Country code","Country"])

# convert html into table data
for row in tables[0].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        alp2 = col[1].text.lower()
        count = col[0].text
        df_cc = df_cc.append({"Country code":alp2,"Country":count}, ignore_index=True)

df_cc
```

```python
dfnew = df.merge(df_cc, on='Country code', how='left')
dfnew
```

```python
writer = pd.ExcelWriter('University Worldwide.xlsx')
# write dataframe to excel
dfnew.to_excel(writer)
# save the excel
writer.save()
```
